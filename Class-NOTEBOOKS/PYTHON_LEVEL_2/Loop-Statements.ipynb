### Looping Statements

#### Break

- With the break statement we can stop the loop before it has looped through all the items:


#### Continue 

- With the continue statement we can stop the current iteration of the loop, and continue with the next:

- skips the current iteration

#### pass
    for/while loops cannot be empty, but if you for some reason have a for loop with no content, put in the pass statement to avoid getting an error

````python
lst = [3,0,9,6,10,20,15,27]
key = 9 
for i in range(0,len(lst)):
    print(f'index i = {i} ')
    if lst[i] == key:
        print(f'index found {i}')
        break
        
````

#### Else with For/While

The else keyword in a for loop specifies a block of code to be executed when the loop is finished:

````python
# Normal usage 

a =10
b= 20
if a>b:
    print("snjknld")
else: #False
    print("else block")
````

````python
#else block is executed when for loop is compeltely executed all iterations
for i in "Kumar":
    print(i ,end="")
    if i=="z":
        break
             
else: #no break #when executed al the iterations
    print(" else block executed")
````

````python
for i in range(1,10):
    print(i) 
    print("hi")
    print("Hello")
    break 
else:
    print("am else block")
````

````python
for i in range(1,30): #stop>start
    print("Hello")
````

````python
while True:
    msg = input("enter the message: ")
    print(msg)
    if msg == "exit":
        break
````

````python
# List Comprehension
numbers = [1,2,3,4,5]
evenFilter = [num**2 for num in numbers if num%2 == 1]
print(evenFilter)
````

````python
# Dictionary Comprehension
num = {1,2,3,4,5}
num_even = {f:f**2 for f in num if f%2== 0}
print(num_even)
````
